```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

def scrape_and_save_article(url, output_file):
    if url is None or url == '':  # Check if the URL is present
        print('URL not provided. Skipping...')
        return
    
    response = requests.get(url)
    
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')
        articles = soup.find_all('article')
        
        content = ''
        
        if len(articles) == 0:  # No articles found
            content = 'No articles found at this URL.'
        else:
            for article in articles:
                title = article.find('h2').get_text() if article.find('h2') else 'No Title'
                paragraphs = article.find_all('p')
                article_content = '\n'.join([p.get_text() for p in paragraphs])
                content += f'Title: {title}\n{article_content}\n\n'
        
        with open(output_file, 'w', encoding='utf-8') as file:
            file.write(content)
            
        print(f'Successfully scraped {url}. Data saved to {output_file}')
    else:
        print(f'Failed to retrieve {url}. Status code: {response.status_code}')
        
        # Save a file with a message indicating no articles were found
        with open(output_file, 'w', encoding='utf-8') as file:
            file.write('No articles found at this URL.')

# Assuming df is your DataFrame containing URLs
# If 'URL' column is missing, you need to replace 'URL' with the correct column name
# List of URLs to scrape
urls = df['URL']

# Scrape and save each article
for i, url in enumerate(urls):
    output_file = f'article_{i + 1}.txt'  # Output file name, you can adjust this as needed
    scrape_and_save_article(url, output_file)
```

```python
import nltk
from nltk.corpus import cmudict
import pandas as pd

# Download the CMU Pronouncing Dictionary
nltk.download('cmudict')

# Load CMU Pronouncing Dictionary
d = cmudict.dict()

# Sample DataFrame 'dfo' with 'Sentence' column
# Assuming 'Sentence' column contains the sentences to be analyzed

# Function to count syllables in a word using CMU Pronouncing Dictionary
def syllable_count(word):
    """Count the number of syllables in a word."""
    try:
        return max([len(list(y for y in x if y[-1].isdigit())) for x in d[word.lower()]])
    except KeyError:
        # If word not found in CMU Pronouncing Dictionary, return 0 syllables
        return 0

# Calculate syllable count for each word in each sentence
def calculate_syllables_per_word(sentence):
    """Calculate average number of syllables per word in a sentence."""
    words = sentence.split()
    syllable_counts = [syllable_count(word) for word in words]
    total_syllables = sum(syllable_counts)
    total_words = len(words)
    if total_words == 0:
        return 0  # Avoid division by zero
    return total_syllables / total_words

# Apply function to calculate syllables per word to each sentence in the DataFrame
dfo['syllables_per_word'] = dfo['Sentence'].apply(calculate_syllables_per_word)

# Calculate the average syllables per word across all sentences
avg_syllables_per_word = dfo['syllables_per_word'].mean()

# Print the average syllables per word
print("Average syllables per word:", avg_syllables_per_word)
```

```python
import nltk
import pandas as pd

# Sample DataFrame 'dfo' with 'Sentence' column
# Assuming 'Sentence' column contains the sentences to be analyzed

# Download the NLTK data for part-of-speech tagging
nltk.download('averaged_perceptron_tagger')

# Function to count personal pronouns in a sentence
def count_personal_pronouns(sentence):
    """Count the occurrences of personal pronouns in a sentence."""
    # Tokenize the sentence into words
    words = nltk.word_tokenize(sentence)
    # Perform part-of-speech tagging to identify personal pronouns
    pos_tags = nltk.pos_tag(words)
    # Count the occurrences of personal pronouns (PRP stands for personal pronoun)
    personal_pronouns_count = sum(1 for word, pos in pos_tags if pos == 'PRP')
    return personal_pronouns_count

# Apply function to count personal pronouns to each sentence in the DataFrame
dfo['personal_pronouns_count'] = dfo['Sentence'].apply(count_personal_pronouns)

```

```python
# Sample DataFrame 'dfo' with 'Sentence' column
# Assuming 'Sentence' column contains the sentences to be analyzed

# Calculate average word length for each sentence
def calculate_avg_word_length(sentence):
    """Calculate average word length in a sentence."""
    words = sentence.split()
    word_lengths = [len(word) for word in words]
    if len(word_lengths) == 0:
        return 0  # Avoid division by zero
    return sum(word_lengths) / len(word_lengths)

# Apply function to calculate average word length to each sentence in the DataFrame
dfo['avg_word_length'] = dfo['Sentence'].apply(calculate_avg_word_length)

```
